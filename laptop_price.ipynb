```python
# Create DataFrame
df = pd.DataFrame(df)

# Count the occurrence of each OS for each Brand
brand_os_counts = df.groupby('Brand')['Operating System'].value_counts().unstack()

# **Increase figure size more significantly**
fig, axes = plt.subplots(1, len(brand_os_counts), figsize=(len(brand_os_counts) * 7, 10))  # Even bigger!

# Ensure axes is iterable even for a single brand
if len(brand_os_counts) == 1:
    axes = [axes]

# **Plot Pie Chart for each Brand**
for i, brand in enumerate(brand_os_counts.index):
    axes[i].pie(brand_os_counts.loc[brand].dropna(), labels=brand_os_counts.loc[brand].dropna().index,
                autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'gold', 'purple'],
                textprops={'fontsize': 18})  # **Even bigger font**
    axes[i].set_title(f'{brand} OS Distribution', fontsize=22)  # **Larger title for better visibility**

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Create DataFrame
df = pd.DataFrame(df)

# Create a pivot table counting occurrences of each OS per brand
brand_os_count = pd.crosstab(df['Brand'], df['Operating System'])

# Display the table
print(brand_os_count)

# Plot a heatmap for better visualization
plt.figure(figsize=(6, 4))
plt.title("Operating System Usage by Brand")
sns.heatmap(brand_os_count, annot=True, cmap="coolwarm", linewidths=0.5, fmt="d")
plt.xlabel("Operating System")
plt.ylabel("Brand")
plt.show()
```

```python
# Create DataFrame
df = pd.DataFrame(df)

# Count how many times each brand uses each OS
brand_os_count = pd.crosstab(df['Brand'], df['Operating System'])

# Print the table
print(brand_os_count)

# Use if-else to check OS for each brand
print("\n Brand and Operating System Analysis:")

for brand in df['Brand'].unique():
    os_used = df[df['Brand'] == brand]['Operating System'].unique()

    print(f"\n {brand} uses: {', '.join(os_used)}")

    # Using if-else conditions
    if 'Windows' in os_used:
        print(f" {brand} supports Windows!")
    elif 'Linux' in os_used:
        print(f" {brand} is Linux-friendly!")
    elif 'MacOS' in os_used:
        print(f" {brand} is an Apple product!")
    elif 'FreeDOS' in os_used:
        print(f" {brand} might require manual OS installation.")
    else:
        print(f" {brand} has an unknown OS.")

# Plot heatmap for better visualization
plt.figure(figsize=(8, 6))
plt.title("Operating System Usage by Brand")
sns.heatmap(brand_os_count, annot=True, cmap="coolwarm", linewidths=0.5, fmt="d")
plt.xlabel("Operating System")
plt.ylabel("Brand")
plt.show()
```
